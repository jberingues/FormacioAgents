### 2.3 Esquemes Pydantic

Definim els models de dades estructurats que usarem:

````python
from pydantic import BaseModel, Field
from typing import List, Optional
from datetime import date

class ComponentToQuote(BaseModel):
    """Component que necessita cotitzaci√≥"""
    part_number: str = Field(description="Codi del component")
    description: str = Field(description="Descripci√≥ t√®cnica")
    quantity: int = Field(description="Quantitat necess√†ria", gt=0)
    manufacturer: str = Field(description="Fabricant")
    priority: str = Field(description="Prioritat: high/medium/low")
    reason: str = Field(description="Per qu√® es cotitza (car, cr√≠tic, etc.)")

class QuoteRequest(BaseModel):
    """Sol¬∑licitud de cotitzaci√≥ (RFQ)"""
    rfq_number: str = Field(description="N√∫mero de RFQ")
    company: str = Field(description="Empresa sol¬∑licitant")
    contact_person: str = Field(description="Persona de contacte")
    contact_email: str = Field(description="Email de contacte")
    components: List[ComponentToQuote] = Field(description="Components a cotitzar")
    notes: Optional[str] = Field(default=None, description="Notes addicionals")

class ComponentQuote(BaseModel):
    """Cotitzaci√≥ d'un component per un prove√Ødor"""
    part_number: str
    supplier: str
    unit_price: float = Field(description="Preu per unitat", gt=0)
    currency: str = Field(default="EUR")
    quantity_available: Optional[int] = Field(default=None, description="Stock disponible")
    lead_time: str = Field(description="Temps de lliurament")
    moq: Optional[int] = Field(default=None, description="Minimum Order Quantity")
    valid_until: Optional[str] = Field(default=None, description="Validesa del preu")
    notes: Optional[str] = Field(default=None, description="Observacions")

class SupplierResponse(BaseModel):
    """Resposta completa d'un prove√Ødor"""
    supplier: str
    contact_email: str
    response_date: str
    rfq_number: str
    quotes: List[ComponentQuote]
    shipping_cost: Optional[float] = Field(default=None)
    shipping_conditions: Optional[str] = Field(default=None)
    payment_terms: Optional[str] = Field(default=None)
    additional_notes: Optional[str] = Field(default=None)

class PriceComparison(BaseModel):
    """Comparativa de preus per un component"""
    part_number: str
    description: str
    quantity_needed: int
    best_quote: ComponentQuote
    all_quotes: List[ComponentQuote]
    price_range: tuple[float, float] = Field(description="(min, max)")
    total_cost_best: float = Field(description="Cost total amb millor oferta")
    savings_vs_worst: float = Field(description="Estalvi vs pitjor oferta")
    recommendation: str
    alerts: List[str] = Field(default_factory=list)

print("‚úÖ Esquemes Pydantic definits!")
print("\nModels disponibles:")
print("  - ComponentToQuote")
print("  - QuoteRequest")
print("  - ComponentQuote")
print("  - SupplierResponse")
print("  - PriceComparison")
````

### 3.1 Tool: Read Excel BOM (reutilitzem de Part 2)

Usarem la mateixa tool que vam crear a la Part 2, per√≤ millorada:

````python
from crewai.tools import tool
import pandas as pd

@tool("Read BOM Excel")
def read_bom_excel(filepath: str) -> str:
    """
    Llegeix un fitxer Excel que cont√© un Bill of Materials (BOM) de JCM Technologies.
    Processa el format espec√≠fic de JCM amb metadata inicial i headers a la fila 7.
    
    Args:
        filepath: Ruta al fitxer Excel (ex: 'data/Bill_of_Materials-TESTBOM.xlsx')
        
    Returns:
        String amb el contingut del BOM en format llegible i detallat
    """
    try:
        # Llegir Excel saltant les primeres 6 files (metadata)
        # La fila 7 (√≠ndex 6) cont√© els headers
        df_bom = pd.read_excel(filepath, skiprows=7)
        
        # Filtrar nom√©s files amb MATERIAL (components reals)
        df_bom = df_bom[df_bom['MATERIAL'].notna()].copy()
        df_bom['MATERIAL'] = df_bom['MATERIAL'].astype(str).str.strip()
        df_bom['MATERIAL'] = df_bom['MATERIAL'].str.replace(r'\.0+$', '', regex=True)
        df_bom = df_bom[df_bom['MATERIAL'].str.match(r'^\d+$', na=False)].copy()
    
        if len(df_bom) == 0:
            return "ERROR: No s'han trobat components v√†lids al BOM"
        
        # Netejar quantitats
        df_bom['Quantity'] = pd.to_numeric(df_bom['Quantity'], errors='coerce').fillna(1).astype(int)
        
        # Generar output detallat
        output = f"BOM REAL DE JCM TECHNOLOGIES (TESTBOM)\n"
        output += f"Llegit de: {filepath}\n\n"
        output += f"‚ïê" * 80 + "\n"
        output += f"RESUM GENERAL\n"
        output += f"‚ïê" * 80 + "\n"
        output += f"Total de components diferents: {len(df_bom)}\n"
        output += f"Total d'unitats necess√†ries: {df_bom['Quantity'].sum()}\n"
        
        # Categoritzar components per tipus
        def get_category(desc):
            if pd.isna(desc):
                return 'Unknown'
            desc_lower = str(desc).lower()
            if 'diode' in desc_lower:
                if 'led' in desc_lower:
                    return 'LED'
                elif 'zener' in desc_lower:
                    return 'Diode-Zener'
                elif 'schottky' in desc_lower:
                    return 'Diode-Schottky'
                else:
                    return 'Diode'
            elif 'cap' in desc_lower or 'capacitor' in desc_lower:
                if 'alum' in desc_lower or 'electrolytic' in desc_lower:
                    return 'Capacitor-Electrolytic'
                return 'Capacitor-Ceramic'
            elif 'res' in desc_lower or 'resistor' in desc_lower:
                return 'Resistor'
            elif 'ic ' in desc_lower or 'mcu' in desc_lower:
                return 'IC'
            elif 'relay' in desc_lower or 'rel√©' in desc_lower:
                return 'Relay'
            elif 'trans' in desc_lower:
                if 'former' in desc_lower:
                    return 'Transformer'
                return 'Transistor'
            elif 'switch' in desc_lower:
                return 'Switch'
            elif 'buzzer' in desc_lower:
                return 'Buzzer'
            elif 'fuse' in desc_lower:
                return 'Fuse/Protection'
            else:
                return 'Other'
        
        df_bom['Category'] = df_bom['Description'].apply(get_category)
        
        output += f"\nComponents per categoria:\n"
        category_counts = df_bom['Category'].value_counts()
        for cat, count in category_counts.items():
            output += f"  - {cat}: {count} components\n"
        
        output += f"\n" + "‚ïê" * 80 + "\n"
        output += f"DETALL DELS COMPONENTS\n"
        output += f"‚ïê" * 80 + "\n\n"
        
        for idx, (_, row) in enumerate(df_bom.iterrows(), 1):
            output += f"Component {idx}. Part Number: {row['MATERIAL']}\n"
            output += f"   Descripci√≥: {row['Description']}\n"
            output += f"   Quantitat: {row['Quantity']} pcs\n"
            output += f"   Categoria: {row['Category']}\n"
            output += f"   {'-' * 76}\n"
        
        return output
        
    except FileNotFoundError:
        return f"ERROR: No s'ha trobat el fitxer {filepath}"
    except KeyError as e:
        return f"ERROR: No s'ha trobat la columna esperada: {e}\n\nEl BOM de JCM ha de tenir columnes: MATERIAL, Description, Quantity"
    except Exception as e:
        return f"ERROR llegint el fitxer: {str(e)}\n\nDetalls: El BOM de JCM t√© un format espec√≠fic amb metadata a les primeres 6 files i headers a la fila 7."

print("‚úÖ Tool 'read_bom_excel' creada (processament format JCM)!")
````

### 5.1 Tool: Authenticate Gmail

Primer creem una tool per autenticar amb Gmail:

````python
import os
from google.auth.transport.requests import Request
from google.oauth2.credentials import Credentials
from google_auth_oauthlib.flow import InstalledAppFlow
from googleapiclient.discovery import build

# Permisos necessaris
SCOPES = [
    'https://www.googleapis.com/auth/gmail.send',      # Enviar emails
    'https://www.googleapis.com/auth/gmail.readonly',  # Llegir emails
    'https://www.googleapis.com/auth/gmail.modify'     # Marcar com llegit
]

def authenticate_gmail():
    """
    Autentica amb Gmail API i retorna el servei.
    
    Returns:
        Gmail API service object
    """
    creds = None
    
    # Si ja existeix token.json, el carrega
    if os.path.exists('token.json'):
        creds = Credentials.from_authorized_user_file('token.json', SCOPES)
    
    # Si no hi ha credencials v√†lides, autentica
    if not creds or not creds.valid:
        if creds and creds.expired and creds.refresh_token:
            print("üîÑ Renovant token...")
            creds.refresh(Request())
        else:
            if not os.path.exists('credentials.json'):
                raise FileNotFoundError(
                    "‚ùå No s'ha trobat credentials.json!\n"
                    "Segueix les instruccions de la secci√≥ 5.1 per crear-lo."
                )
            
            print("üîê Primera autenticaci√≥...")
            print("   S'obrir√† un navegador per autenticar-te.")
            flow = InstalledAppFlow.from_client_secrets_file(
                'credentials.json', SCOPES
            )
            creds = flow.run_local_server(port=0)
        
        # Guardar credencials per seg√ºents vegades
        with open('token.json', 'w') as token:
            token.write(creds.to_json())
        print("‚úÖ Token guardat a token.json")
    
    # Crear servei Gmail
    service = build('gmail', 'v1', credentials=creds)
    
    # Obtenir info del compte
    profile = service.users().getProfile(userId='me').execute()
    print(f"‚úÖ Autenticat com: {profile['emailAddress']}")
    
    return service

print("‚úÖ Funci√≥ authenticate_gmail() definida")
````

### 5.3 Tool: Send Email

Ara creem la tool per enviar emails:

````python
from email.mime.text import MIMEText
import base64
from crewai.tools import tool

@tool("Send Email via Gmail")
def send_email_gmail(
    to_email: str,
    subject: str,
    body: str
) -> str:
    """
    Envia un email via Gmail API.
    
    Args:
        to_email: Email destinatari (ex: 'colleague@jcm-tech.com')
        subject: Assumpte de l'email
        body: Cos del missatge (text pla)
        
    Returns:
        Confirmaci√≥ de l'enviament amb ID del missatge
    """
    try:
        # Autenticar
        service = authenticate_gmail()
        
        # Crear missatge
        message = MIMEText(body)
        message['to'] = to_email
        message['subject'] = subject
        
        # Codificar en base64
        raw_message = base64.urlsafe_b64encode(message.as_bytes()).decode('utf-8')
        
        # Enviar
        sent_message = service.users().messages().send(
            userId='me',
            body={'raw': raw_message}
        ).execute()
        
        result = f"""
‚úÖ EMAIL ENVIAT AMB √àXIT!

Destinatari: {to_email}
Assumpte: {subject}
ID del missatge: {sent_message['id']}

L'email ha estat enviat correctament via Gmail.
"""
        return result
        
    except Exception as e:
        return f"‚ùå ERROR enviant email: {str(e)}"

print("‚úÖ Tool 'send_email_gmail' creada!")
````

### 5.4 Tool: Fetch Emails

Tool per llegir emails rebuts:

````python
@tool("Fetch Emails from Gmail")
def fetch_emails_gmail(
    query: str = "is:unread",
    max_results: int = 10
) -> str:
    """
    Llegeix emails de la inbox de Gmail.
    
    Args:
        query: Cerca Gmail (ex: 'is:unread subject:RFQ', 'from:colleague@company.com')
        max_results: M√†xim d'emails a retornar (default: 10)
        
    Returns:
        Llista d'emails amb: From, Subject, Date, Body
    """
    try:
        # Autenticar
        service = authenticate_gmail()
        
        # Cercar missatges
        results = service.users().messages().list(
            userId='me',
            q=query,
            maxResults=max_results
        ).execute()
        
        messages = results.get('messages', [])
        
        if not messages:
            return f"‚ÑπÔ∏è No s'han trobat emails que coincideixin amb: '{query}'"
        
        output = f"üìß S'han trobat {len(messages)} email(s):\n\n"
        output += "=" * 80 + "\n\n"
        
        for idx, msg in enumerate(messages, 1):
            # Obtenir detalls del missatge
            msg_data = service.users().messages().get(
                userId='me',
                id=msg['id'],
                format='full'
            ).execute()
            
            # Extreure headers
            headers = msg_data['payload']['headers']
            subject = next((h['value'] for h in headers if h['name'] == 'Subject'), 'Sin asunto')
            from_email = next((h['value'] for h in headers if h['name'] == 'From'), 'Desconocido')
            date = next((h['value'] for h in headers if h['name'] == 'Date'), 'Fecha desconocida')
            
            # Extreure cos del missatge
            body = ""
            if 'parts' in msg_data['payload']:
                for part in msg_data['payload']['parts']:
                    if part['mimeType'] == 'text/plain':
                        if 'data' in part['body']:
                            body = base64.urlsafe_b64decode(
                                part['body']['data']
                            ).decode('utf-8')
                            break
            else:
                if 'data' in msg_data['payload']['body']:
                    body = base64.urlsafe_b64decode(
                        msg_data['payload']['body']['data']
                    ).decode('utf-8')
            
            output += f"EMAIL {idx}\n"
            output += "-" * 80 + "\n"
            output += f"De: {from_email}\n"
            output += f"Assumpte: {subject}\n"
            output += f"Data: {date}\n"
            output += f"ID: {msg['id']}\n"
            output += "-" * 80 + "\n"
            output += f"COS DEL MISSATGE:\n\n{body[:1000]}"  # Limitar a 1000 car√†cters
            if len(body) > 1000:
                output += "\n\n[... missatge truncat ...]\n"
            output += "\n" + "=" * 80 + "\n\n"
        
        return output
        
    except Exception as e:
        return f"‚ùå ERROR llegint emails: {str(e)}"

print("‚úÖ Tool 'fetch_emails_gmail' creada!")
````

## 7. Agent 3: Email Parser ‚≠ê

**Aquesta √©s la part m√©s interessant!** L'agent que parseja emails reals amb formats diversos.

### 7.1 Per qu√® aquesta √©s la part clau?

**El desafiament:**
- ‚ùå Els emails NO estan estructurats
- ‚ùå Cada prove√Ødor usa un format diferent
- ‚ùå Pot faltar informaci√≥
- ‚ùå Poden haver errors tipogr√†fics
- ‚ùå Pot estar en idiomes diversos

**El poder dels LLMs:**
- ‚úÖ Entenen context i significat
- ‚úÖ Poden extreure info de text no estructurat
- ‚úÖ Gestionen formats diversos
- ‚úÖ Normalitzen dades autom√†ticament
- ‚úÖ Toleren errors i inconsist√®ncies

**Aix√≤ NO es pot fer amb regex o parsers tradicionals!**

### 7.2 Crear l'Agent Email Parser

````python
agent_email_parser = Agent(
    role="Expert en Parsing d'Emails de Cotitzacions",
    goal="Extreure preus, stock i condicions d'emails de prove√Ødors amb formats diversos",
    backstory="""
    Ets un especialista en processament de llenguatge natural amb 10 anys d'experi√®ncia
    processant emails comercials.
    
    La teva expertesa √©s:
    - Llegir emails de prove√Ødors en formats MOLT diversos
    - Extreure informaci√≥ clau: preus, stock, lead times, condicions
    - Normalitzar dades (convertir monedes, unitats, dates)
    - Gestionar informaci√≥ incompleta o ambigua
    - Detectar warnings o condicions especials
    
    Pots processar:
    - Taules ben formatades
    - Text lliure conversacional
    - Formats ca√≤tics o inconsistents
    - M√∫ltiples idiomes (catal√†, castell√†, angl√®s)
    
    Tens acc√©s a eines per llegir emails de Gmail.
    
    MOLT IMPORTANT:
    - Si no trobes una dada, marca-la com "No especificat" (NO inventes!)
    - Si hi ha ambig√ºitat, indica-ho a les notes
    - Converteix sempre a EUR si cal
    - Normalitza lead times a format est√†ndard (ex: "3-5 dies", "2 setmanes")
    
    El teu output ha de ser structured amb Pydantic per poder-lo processar despr√©s.
    """,
    tools=[fetch_emails_gmail],
    llm=OPENAI_MODEL,
    verbose=True
)

print("‚úÖ Agent 'Email Parser' creat!")
````

### 7.3 Task: Parse Email Response

````python
task_parse_emails = Task(
    description=f"""
    Llegeix les 3 respostes als RFQs que hem enviat als voluntaris.
    
    Passos:
    1. Usa la tool 'Fetch Emails from Gmail' per llegir emails recents
       Cerca: "subject:cotitzaci√≥ OR subject:RFQ newer_than:1d"
    
    2. Identifica les 3 respostes dels voluntaris:
       - DigiKey: {VOLUNTARI_DIGIKEY}
       - Mouser: {VOLUNTARI_MOUSER}
       - Farnell: {VOLUNTARI_FARNELL}
    
    3. Per CADA email rebut, extreu:
       - Nom del prove√Ødor (DigiKey/Mouser/Farnell)
       - Per CADA component mencionat:
         * Part Number (exacte)
         * Preu unitari (convertir a EUR si cal)
         * Stock disponible
         * Lead time
         * MOQ si est√† especificat
         * Notes especials
       - Informaci√≥ general:
         * Cost d'enviament
         * Condicions de pagament
         * Validesa de la cotitzaci√≥
    
    CR√çTIC:
    - Gestiona formats diversos (taula, text lliure, ca√≤tic)
    - Si una dada no est√†, marca-la com "No especificat" (NO inventes!)
    - Normalitza preus (sempre en EUR)
    - Processa les 3 respostes, no nom√©s una
    """,
    expected_output="""
    Dades extretes de les 3 respostes en format estructurat:
    
    ## Resposta 1: DigiKey
    - Email: {VOLUNTARI_DIGIKEY}
    - Data: [data]
    
    ### Components:
    - Part Number: [codi], Preu: ‚Ç¨X.XX, Stock: [info], Lead time: [temps]
    [... m√©s components ...]
    
    ### Condicions:
    - Enviament: ‚Ç¨XX
    - Pagament: [terminis]
    - Validesa: [data]
    
    ## Resposta 2: Mouser
    [... mateix format ...]
    
    ## Resposta 3: Farnell
    [... mateix format ...]
    """,
    agent=agent_email_parser,
    tools=[fetch_emails_gmail]
)

print("‚úÖ Task 'Parse Emails' creada per processar les 3 respostes!")
````

### 7.4 Executar Email Parser

**‚ö†Ô∏è Assegura't que:**
1. Han arribat els correus dels voluntaris.

````python
# Crew temporal per provar el parser
crew_parse = Crew(
    agents=[agent_email_parser],
    tasks=[task_parse_emails],  # ‚Üê Canviat de task_parse_email a task_parse_emails
    process=Process.sequential,
    verbose=False  # ‚Üê Desactivar verbose per evitar RecursionError
)

print("üöÄ Executant Email Parser...\n")
print("=" * 70)
print(f"\nüì¨ Buscant respostes de:")
print(f"   ‚Ä¢ DigiKey:  {VOLUNTARI_DIGIKEY}")
print(f"   ‚Ä¢ Mouser:   {VOLUNTARI_MOUSER}")
print(f"   ‚Ä¢ Farnell:  {VOLUNTARI_FARNELL}")
print("\n‚è≥ Llegint i processant emails...\n")
print("=" * 70)

result_parse = crew_parse.kickoff()

print("\n" + "=" * 70)
print("‚úÖ Parsing completat!")
print("=" * 70)

# Verificar si √©s Pydantic
if hasattr(result_parse, 'pydantic') and result_parse.pydantic:
    parsed_data = result_parse.pydantic
    
    print("\nüìã DADES ESTRUCTURADES (Pydantic):\n")
    print(f"Prove√Ødor: {parsed_data.supplier_name}")  # ‚Üê Corregit
    print(f"Email: {parsed_data.supplier_email}")
    print(f"Data: {parsed_data.response_date}")
    
    print(f"\nComponents cotitzats: {len(parsed_data.components)}")  # ‚Üê Corregit
    for component in parsed_data.components:  # ‚Üê Corregit
        print(f"\n  ‚Ä¢ {component.part_number}")
        if component.description:
            print(f"    Descripci√≥: {component.description}")
        print(f"    Preu: ‚Ç¨{component.unit_price:.2f}")
        print(f"    Stock: {component.stock}")
        print(f"    Lead time: {component.lead_time}")
        if component.moq:
            print(f"    MOQ: {component.moq}")
        if component.notes:
            print(f"    Notes: {component.notes}")
    
    print(f"\nüì¶ Condicions:")
    print(f"  ‚Ä¢ Enviament: {parsed_data.shipping_cost}")
    print(f"  ‚Ä¢ Pagament: {parsed_data.payment_terms}")
    print(f"  ‚Ä¢ Validesa: {parsed_data.quote_validity}")
    if parsed_data.additional_notes:
        print(f"  ‚Ä¢ Notes: {parsed_data.additional_notes}")
else:
    # Si no √©s Pydantic, mostrar text pla
    print("\nüìä DADES EXTRETES:\n")
    print(result_parse)
````

### 8.1 Agent Price Comparator

````python
agent_price_comparator = Agent(
    role="Expert en Comparaci√≥ de Preus i Condicions",
    goal="Comparar cotitzacions de m√∫ltiples prove√Ødors i identificar les millors ofertes",
    backstory="""
    Ets un analista de procurement amb 15 anys d'experi√®ncia a JCM Technologies.
    
    La teva expertesa √©s:
    - Comparar preus de m√∫ltiples prove√Ødors
    - Avaluar NO nom√©s el preu, sin√≥ tamb√©:
      * Disponibilitat de stock
      * Lead times
      * Condicions de pagament
      * Costos d'enviament
      * Fiabilitat del prove√Ødor
    
    - Identificar la millor oferta per cada component
    - Calcular costos totals (preu + enviament)
    - Detectar alertes:
      * Stock baix
      * Lead times llargs (>2 setmanes)
      * Preus anormalment alts o baixos
      * MOQ problem√†tics
    
    - Generar recomanacions pr√†ctiques
    
    El teu an√†lisi sempre considera el COST TOTAL, no nom√©s el preu unitari.
    
    Per exemple:
    - Component ‚Ç¨3.00 + enviament gratis > Component ‚Ç¨2.50 + enviament ‚Ç¨30
    - Stock immediat > Preu 10% m√©s baix amb 6 setmanes lead time
    """,
    tools=[],  # No necessita tools, treballa amb dades de context
    llm=OPENAI_MODEL,
    verbose=True
)

print("‚úÖ Agent 'Price Comparator' creat!")
````

### 8.2 Task: Compare Prices

````python
task_compare_prices = Task(
    description="""
    Analitza les cotitzacions extretes dels emails i genera una comparativa completa.
    
    Basant-te en les dades de la task anterior (context):
    
    1. Per CADA component:
       - Compara preus unitaris
       - Avalua disponibilitat de stock
       - Compara lead times
       - Identifica la millor oferta
       - Calcula cost total: (preu unitari √ó quantitat) + proporci√≥ enviament
    
    2. Detecta alertes:
       - Components amb stock <10 unitats
       - Lead times >3 setmanes
       - Preus >50% m√©s cars que la mitjana
       - MOQ superior a la quantitat necess√†ria
    
    3. Genera recomanacions:
       - Millor prove√Ødor per cada component
       - Possibles estalvis
       - Riscos a considerar
       - Alternativa si millor oferta no disponible
    
    4. Calcula cost total del projecte amb:
       - Millor opci√≥ per cada component
       - Pitjor opci√≥ (per comparar estalvi)
    """,
    expected_output="""
    Informe de comparativa amb aquesta estructura:
    
    ## Resum Executiu
    - Cost total amb millors ofertes: ‚Ç¨XXX.XX
    - Cost total amb pitjors ofertes: ‚Ç¨XXX.XX
    - Estalvi potencial: ‚Ç¨XX.XX (XX%)
    - Components analitzats: [n√∫mero]
    - Alertes detectades: [n√∫mero]
    
    ## Comparativa per Component
    
    ### [Part Number]
    - Quantitat necess√†ria: X unitats
    - Prove√Ødors comparats: [llista]
    - Millor oferta: [Prove√Ødor] - ‚Ç¨X.XX/ud (total: ‚Ç¨XX.XX)
    - Difer√®ncia vs pitjor: ‚Ç¨X.XX (XX%)
    - Stock: [info]
    - Lead time: [temps]
    - ‚ö†Ô∏è Alertes: [si n'hi ha]
    - üí° Recomanaci√≥: [text]
    
    [... repetir per cada component ...]
    
    ## Recomanacions Finals
    1. [Recomanaci√≥ priorit√†ria]
    2. [Recomanaci√≥ secund√†ria]
    3. [Consideracions addicionals]
    """,
    agent=agent_price_comparator,
    context=[task_parse_emails]  # Rep les dades parsejades
)

print("‚úÖ Task 'Compare Prices' creada!")
````
