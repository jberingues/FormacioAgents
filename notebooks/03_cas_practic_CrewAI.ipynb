### 2.3 Esquemes Pydantic

Definim els models de dades estructurats que usarem:

````python
from pydantic import BaseModel, Field
from typing import List, Optional
from datetime import date

class ComponentToQuote(BaseModel):
    """Component que necessita cotitzaciÃ³"""
    part_number: str = Field(description="Codi del component")
    description: str = Field(description="DescripciÃ³ tÃ¨cnica")
    quantity: int = Field(description="Quantitat necessÃ ria", gt=0)
    manufacturer: str = Field(description="Fabricant")
    category: str = Field(description="Categoria")

class QuoteRequest(BaseModel):
    """SolÂ·licitud de cotitzaciÃ³ (RFQ)"""
    rfq_number: str = Field(description="NÃºmero de RFQ")
    company: str = Field(description="Empresa solÂ·licitant")
    contact_person: str = Field(description="Persona de contacte")
    contact_email: str = Field(description="Email de contacte")
    components: List[ComponentToQuote] = Field(description="Components a cotitzar")
    notes: Optional[str] = Field(default=None, description="Notes addicionals")

class ComponentQuote(BaseModel):
    """CotitzaciÃ³ d'un component per un proveÃ¯dor"""
    part_number: str
    supplier: str
    unit_price: float = Field(description="Preu per unitat", gt=0)
    currency: str = Field(default="EUR")
    quantity_available: Optional[int] = Field(default=None, description="Stock disponible")
    lead_time: str = Field(description="Temps de lliurament")
    moq: Optional[int] = Field(default=None, description="Minimum Order Quantity")
    valid_until: Optional[str] = Field(default=None, description="Validesa del preu")
    notes: Optional[str] = Field(default=None, description="Observacions")

class SupplierResponse(BaseModel):
    """Resposta completa d'un proveÃ¯dor"""
    supplier: str
    contact_email: str
    response_date: str
    rfq_number: str
    quotes: List[ComponentQuote]
    shipping_cost: Optional[float] = Field(default=None)
    shipping_conditions: Optional[str] = Field(default=None)
    payment_terms: Optional[str] = Field(default=None)
    additional_notes: Optional[str] = Field(default=None)

class PriceComparison(BaseModel):
    """Comparativa de preus per un component"""
    part_number: str
    description: str
    quantity_needed: int
    best_quote: ComponentQuote
    all_quotes: List[ComponentQuote]
    price_range: tuple[float, float] = Field(description="(min, max)")
    total_cost_best: float = Field(description="Cost total amb millor oferta")
    savings_vs_worst: float = Field(description="Estalvi vs pitjor oferta")
    recommendation: str
    alerts: List[str] = Field(default_factory=list)

print("âœ… Esquemes Pydantic definits!")
print("\nModels disponibles:")
print("  - ComponentToQuote")
print("  - QuoteRequest")
print("  - ComponentQuote")
print("  - SupplierResponse")
print("  - PriceComparison")
````

## 3. Agent 1: BOM Analyzer Expert

### 3.1 Tool: Read Excel BOM (reutilitzem de Part 2)

Usarem la mateixa tool que vam crear a la Part 2, perÃ² millorada:

````python
from crewai.tools import tool
import pandas as pd
from typing import List
import json

@tool("Read BOM Excel")
def read_bom_excel(filepath: str) -> str:
    """
    Llegeix un fitxer Excel que contÃ© un Bill of Materials (BOM) de JCM Technologies.
    Processa el format especÃ­fic de JCM amb metadata inicial i headers a la fila 7.
    Retorna una llista de ComponentToQuote en format JSON.
    
    Args:
        filepath: Ruta al fitxer Excel (ex: 'data/Bill_of_Materials-TESTBOM.xlsx')
        
    Returns:
        JSON string amb llista de ComponentToQuote objects
    """
    try:
        # Llegir Excel saltant les primeres 6 files (metadata)
        # La fila 7 (Ã­ndex 6) contÃ© els headers
        df_bom = pd.read_excel(filepath, skiprows=7)
        
        # Filtrar nomÃ©s files amb MATERIAL (components reals)
        df_bom = df_bom[df_bom['MATERIAL'].notna()].copy()
        df_bom['MATERIAL'] = df_bom['MATERIAL'].astype(str).str.strip()
        df_bom['MATERIAL'] = df_bom['MATERIAL'].str.replace(r'\.0+$', '', regex=True)
        df_bom = df_bom[df_bom['MATERIAL'].str.match(r'^\d+$', na=False)].copy()
    
        if len(df_bom) == 0:
            return json.dumps([])
        
        # Netejar quantitats
        df_bom['Quantity'] = pd.to_numeric(df_bom['Quantity'], errors='coerce').fillna(1).astype(int)
        
        # Categoritzar components per tipus
        def get_category(desc):
            if pd.isna(desc):
                return 'Unknown'
            desc_lower = str(desc).lower()
            if 'diode' in desc_lower:
                if 'led' in desc_lower:
                    return 'LED'
                elif 'zener' in desc_lower:
                    return 'Diode-Zener'
                elif 'schottky' in desc_lower:
                    return 'Diode-Schottky'
                else:
                    return 'Diode'
            elif 'cap' in desc_lower or 'capacitor' in desc_lower:
                if 'alum' in desc_lower or 'electrolytic' in desc_lower:
                    return 'Capacitor-Electrolytic'
                return 'Capacitor-Ceramic'
            elif 'res' in desc_lower or 'resistor' in desc_lower:
                return 'Resistor'
            elif 'ic ' in desc_lower or 'mcu' in desc_lower:
                return 'IC'
            elif 'relay' in desc_lower or 'relÃ©' in desc_lower:
                return 'Relay'
            elif 'trans' in desc_lower:
                if 'former' in desc_lower:
                    return 'Transformer'
                return 'Transistor'
            elif 'switch' in desc_lower:
                return 'Switch'
            elif 'buzzer' in desc_lower:
                return 'Buzzer'
            elif 'fuse' in desc_lower:
                return 'Fuse/Protection'
            else:
                return 'Other'
        
        df_bom['Category'] = df_bom['Description'].apply(get_category)
        
        # Crear llista de ComponentToQuote
        components = []
        for _, row in df_bom.iterrows():
            # Obtenir manufacturer si existeix, sinÃ³ "Unknown"
            manufacturer = "Unknown"
            if 'Manufacturer' in df_bom.columns and pd.notna(row.get('Manufacturer')):
                manufacturer = str(row['Manufacturer'])
            elif 'Fabricant' in df_bom.columns and pd.notna(row.get('Fabricant')):
                manufacturer = str(row['Fabricant'])
            
            component = ComponentToQuote(
                part_number=str(row['MATERIAL']),
                description=str(row['Description']) if pd.notna(row['Description']) else "No description",
                quantity=int(row['Quantity']),
                manufacturer=manufacturer,
                category=row['Category']
            )
            components.append(component)
        
        # Retornar com a JSON string (serialitzant els objectes Pydantic)
        return json.dumps([comp.model_dump() for comp in components], indent=2, ensure_ascii=False)
        
    except FileNotFoundError:
        return json.dumps({"error": f"No s'ha trobat el fitxer {filepath}"})
    except KeyError as e:
        return json.dumps({"error": f"No s'ha trobat la columna esperada: {e}"})
    except Exception as e:
        return json.dumps({"error": f"ERROR llegint el fitxer: {str(e)}"})

print("âœ… Tool 'read_bom_excel' creada (retorna ComponentToQuote en format JSON)!")
````

### 5.1 Tool: Authenticate Gmail

Primer creem una tool per autenticar amb Gmail:

````python
import os
from google.auth.transport.requests import Request
from google.oauth2.credentials import Credentials
from google_auth_oauthlib.flow import InstalledAppFlow
from googleapiclient.discovery import build

# Permisos necessaris
SCOPES = [
    'https://www.googleapis.com/auth/gmail.send',      # Enviar emails
    'https://www.googleapis.com/auth/gmail.readonly',  # Llegir emails
    'https://www.googleapis.com/auth/gmail.modify'     # Marcar com llegit
]

def authenticate_gmail():
    """
    Autentica amb Gmail API i retorna el servei.
    
    Returns:
        Gmail API service object
    """
    creds = None
    
    # Si ja existeix token.json, el carrega
    if os.path.exists('token.json'):
        creds = Credentials.from_authorized_user_file('token.json', SCOPES)
    
    # Si no hi ha credencials vÃ lides, autentica
    if not creds or not creds.valid:
        if creds and creds.expired and creds.refresh_token:
            print("ğŸ”„ Renovant token...")
            creds.refresh(Request())
        else:
            if not os.path.exists('credentials.json'):
                raise FileNotFoundError(
                    "âŒ No s'ha trobat credentials.json!\n"
                    "Segueix les instruccions de la secciÃ³ 5.1 per crear-lo."
                )
            
            print("ğŸ” Primera autenticaciÃ³...")
            print("   S'obrirÃ  un navegador per autenticar-te.")
            flow = InstalledAppFlow.from_client_secrets_file(
                'credentials.json', SCOPES
            )
            creds = flow.run_local_server(port=0)
        
        # Guardar credencials per segÃ¼ents vegades
        with open('token.json', 'w') as token:
            token.write(creds.to_json())
        print("âœ… Token guardat a token.json")
    
    # Crear servei Gmail
    service = build('gmail', 'v1', credentials=creds)
    
    # Obtenir info del compte
    profile = service.users().getProfile(userId='me').execute()
    print(f"âœ… Autenticat com: {profile['emailAddress']}")
    
    return service

print("âœ… FunciÃ³ authenticate_gmail() definida")
````

````python
from email.mime.text import MIMEText
import base64
from crewai.tools import tool

@tool("Send Email via Gmail")
def send_email_gmail(
    to_email: str,
    subject: str,
    body: str
) -> str:
    """
    Envia un email via Gmail API.
    
    Args:
        to_email: Email destinatari (ex: 'colleague@jcm-tech.com')
        subject: Assumpte de l'email
        body: Cos del missatge (text pla)
        
    Returns:
        ConfirmaciÃ³ de l'enviament amb ID del missatge
    """
    try:
        # Usar el servei global ja autenticat
        global gmail_service

        # Crear missatge
        message = MIMEText(body)
        message['to'] = to_email
        message['subject'] = subject
        
        # Codificar en base64
        raw_message = base64.urlsafe_b64encode(message.as_bytes()).decode('utf-8')
        
        # Enviar
        sent_message = service.users().messages().send(
            userId='me',
            body={'raw': raw_message}
        ).execute()
        
        return f"""âœ… Email enviat a {to_email}"""
        
    except Exception as e:
        return f"âŒ ERROR enviant email: {str(e)}"

print("âœ… Tool 'send_email_gmail' creada!")
````

### 5.4 Tool: Fetch Emails

Tool per llegir emails rebuts:

````python
@tool("Fetch Emails from Gmail")
def fetch_emails_gmail(
    query: str = "is:unread",
    max_results: int = 10
) -> str:
    """
    Llegeix emails de la inbox de Gmail.
    
    Args:
        query: Cerca Gmail (ex: 'is:unread subject:RFQ', 'from:colleague@company.com')
        max_results: MÃ xim d'emails a retornar (default: 10)
        
    Returns:
        Llista d'emails amb: From, Subject, Date, Body
    """
    try:
        # Usar el servei global ja autenticat
        global gmail_service

        # Cercar missatges
        results = service.users().messages().list(
            userId='me',
            q=query,
            maxResults=max_results
        ).execute()
        
        messages = results.get('messages', [])
        
        if not messages:
            return f"â„¹ï¸ No s'han trobat emails que coincideixin amb: '{query}'"
        
        output = f"ğŸ“§ S'han trobat {len(messages)} email(s):\n\n"
        output += "=" * 80 + "\n\n"
        
        for idx, msg in enumerate(messages, 1):
            # Obtenir detalls del missatge
            msg_data = service.users().messages().get(
                userId='me',
                id=msg['id'],
                format='full'
            ).execute()
            
            # Extreure headers
            headers = msg_data['payload']['headers']
            subject = next((h['value'] for h in headers if h['name'] == 'Subject'), 'Sin asunto')
            from_email = next((h['value'] for h in headers if h['name'] == 'From'), 'Desconocido')
            date = next((h['value'] for h in headers if h['name'] == 'Date'), 'Fecha desconocida')
            
            # Extreure cos del missatge
            body = ""
            if 'parts' in msg_data['payload']:
                for part in msg_data['payload']['parts']:
                    if part['mimeType'] == 'text/plain':
                        if 'data' in part['body']:
                            body = base64.urlsafe_b64decode(
                                part['body']['data']
                            ).decode('utf-8')
                            break
            else:
                if 'data' in msg_data['payload']['body']:
                    body = base64.urlsafe_b64decode(
                        msg_data['payload']['body']['data']
                    ).decode('utf-8')
            
            output += f"EMAIL {idx}\n"
            output += "-" * 80 + "\n"
            output += f"De: {from_email}\n"
            output += f"Assumpte: {subject}\n"
            output += f"Data: {date}\n"
            output += f"ID: {msg['id']}\n"
            output += "-" * 80 + "\n"
            output += f"COS DEL MISSATGE:\n\n{body[:1000]}"  # Limitar a 1000 carÃ cters
            if len(body) > 1000:
                output += "\n\n[... missatge truncat ...]\n"
            output += "\n" + "=" * 80 + "\n\n"
        
        return output
        
    except Exception as e:
        return f"âŒ ERROR llegint emails: {str(e)}"

print("âœ… Tool 'fetch_emails_gmail' creada!")
````
