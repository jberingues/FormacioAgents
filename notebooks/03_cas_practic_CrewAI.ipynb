### 2.2 Esquemes Pydantic

Definim els models de dades estructurats que usarem:

````python
from pydantic import BaseModel, Field
from typing import List, Optional
from datetime import date

class ComponentToQuote(BaseModel):
    """Component que necessita cotitzaci√≥"""
    codi_JCM: str = Field(description="Codi JCM")
    description: str = Field(description="Descripci√≥ t√®cnica")
    quantity: int = Field(description="Quantitat necess√†ria", gt=0)
    manufacturer: str = Field(description="Fabricant")
    part_number: str = Field(description="Refer√®ncia del component del fabricant")
    category: str = Field(description="Categoria")

class ComponentToQuoteList(BaseModel):
    items: List[ComponentToQuote]

class BOMAnalysis(BaseModel):
    """Resultat de l'an√†lisi del BOM"""
    components_to_quote: List[ComponentToQuote] = Field(description="Llista de tots els components que necessiten ser cotitzats")
    components_by_supplier: dict[str, List[str]] = Field( description="Codi JCM agrupats per prove√Ødor")
    summary: str = Field(description="Descripci√≥ de l'an√†lilsis")
    total_components: int = Field(description="N√∫mero de components a cotitzar")

class RFQEmail(BaseModel):
    supplier_name: str = Field(description="Nom del prove√Ødor (DigiKey, Mouser, Farnell)")
    to_email: str = Field(description="Email destinatari del prove√Ødor")
    subject: str = Field(description="Assumpte de l'email")
    body: str = Field(description="Cos de l'email (text pla)")
    rfq_number: str = Field(description="N√∫mero d'RFQ utilitzat")
    from_email: str = Field(description="Email remitent utilitzat")

class RFQEmails(BaseModel):
    """Emails RFQ generats per prove√Ødor (ready-to-send)"""
    emails: List[RFQEmail] = Field(description="Llista d'emails RFQ a enviar")
    total_emails: int = Field(description="N√∫mero total d'emails generats")

class GmailEmail(BaseModel):
    message_id: str = Field(description="ID intern del missatge Gmail")
    from_email: str = Field(description="Remitent")
    subject: str = Field(description="Assumpte")
    date: str = Field(description="Data del correu (string Gmail)")
    snippet: str = Field(description="Snippet curt de Gmail")
    body: str = Field(description="Cos complet del correu (text/plain preferent)")

class GmailInbox(BaseModel):
    emails: List[GmailEmail] = Field(description="Llista de correus llegits")
    total_emails: int = Field(description="Total de correus retornats")

class ParsedQuoteLine(BaseModel):
    supplier: str = Field(description="Prove√Ødor (DigiKey, Mouser, Farnell)")
    codi_JCM: str = Field(description="Codi JCM del component (si es pot identificar)")
    unit_price: Optional[float] = Field(default=None, description="Preu unitari (EUR)")
    currency: Optional[str] = Field(default="EUR", description="Moneda")
    stock: Optional[int] = Field(default=None, description="Stock disponible")
    lead_time_days: Optional[int] = Field(default=None, description="Lead time en dies")
    stock_text: str | None = Field(default=None, description="Stock tal com apareix al correu (ex: 'stock ok', 'alt', '1000+')")
    lead_time_text: str | None = Field(default=None, description="Lead time tal com apareix al correu (ex: '24/48h', '1 setmana', '3-5 dies')")
    moq: Optional[int] = Field(default=None, description="MOQ")
    notes: Optional[str] = Field(default=None, description="Notes / ambig√ºitats / condicions")

class ParsedSupplierResponse(BaseModel):
    supplier: str = Field(description="Prove√Ødor detectat")
    from_email: str = Field(description="Email remitent")
    subject: str = Field(description="Subject del correu")
    date: str = Field(description="Data del correu")
    lines: List[ParsedQuoteLine] = Field(description="L√≠nies parsejades")
    shipping_cost: Optional[str] = Field(default=None, description="Cost d'enviament si existeix")
    payment_terms: Optional[str] = Field(default=None, description="Condicions de pagament si existeixen")
    quote_validity: Optional[str] = Field(default=None, description="Validesa de la cotitzaci√≥ si existeix")
    additional_notes: Optional[str] = Field(default=None, description="Altres notes")

class ParsedQuotes(BaseModel):
    responses: List[ParsedSupplierResponse] = Field(description="Respostes parsejades dels prove√Ødors")
    total_responses: int = Field(description="Nombre total de respostes")

class BestOffer(BaseModel):
    supplier: str = Field(description="Prove√Ødor escollit com a millor oferta")
    unit_price: float = Field(description="Preu unitari escollit (EUR)")
    currency: str = Field(default="EUR", description="Moneda")
    stock: Optional[int] = Field(default=None, description="Stock num√®ric si existeix")
    stock_text: Optional[str] = Field(default=None, description="Text stock si existeix (ok/alt/normal/...)")
    lead_time_days: Optional[int] = Field(default=None, description="Lead time en dies si es pot inferir")
    lead_time_text: Optional[str] = Field(default=None, description="Lead time en format text")
    moq: Optional[int] = Field(default=None, description="MOQ si existeix")
    notes: Optional[str] = Field(default=None, description="Notes addicionals")

class ValuatedLine(BaseModel):
    codi_JCM: str = Field(description="Codi JCM del component")
    description: str = Field(description="Descripci√≥ del BOM")
    manufacturer: str = Field(description="Fabricant")
    part_number: str = Field(description="Part number")
    quantity: int = Field(description="Quantitat del BOM")
    best_offer: Optional[BestOffer] = Field(default=None, description="Millor oferta trobada")
    line_total_eur: Optional[float] = Field(default=None, description="Total l√≠nia en EUR si hi ha oferta")
    quoted_by: List[str] = Field(default_factory=list, description="Llista de prove√Ødors que han cotitzat aquesta l√≠nia")
    missing_quote: bool = Field(default=False, description="True si no hi ha cap cotitzaci√≥ per aquest component")

class SupplierTotal(BaseModel):
    supplier: str
    subtotal_eur: float
    num_lines: int

class ValuatedBOM(BaseModel):
    lines: List[ValuatedLine]
    total_lines: int
    total_valuated_lines: int
    total_missing_lines: int
    grand_total_eur: float
    totals_by_supplier: List[SupplierTotal]



print("‚úÖ Esquemes Pydantic definits!")
````

### 3.1 Tool: Read Excel BOM (reutilitzem de Part 2)

Usarem la mateixa tool que vam crear a la Part 2, per√≤ millorada, a m√©s a m√©s de llegir la llista de materials, tamb√© els agrupa per categories segons la descripci√≥:

````python
from crewai.tools import tool
import pandas as pd
from typing import List
import json

@tool("Read BOM Excel")
def read_bom_excel(filepath: str) -> str:
    """
    Llegeix un fitxer Excel que cont√© un Bill of Materials (BOM) de JCM Technologies.
    Processa el format espec√≠fic de JCM amb metadata inicial i headers a la fila 7.
    Retorna una llista de ComponentToQuote en format JSON.
    
    Args:
        filepath: Ruta al fitxer Excel (ex: 'data/Bill_of_Materials-TESTBOM.xlsx')
        
    Returns:
        JSON string amb llista de ComponentToQuote objects
    """
    try:
        print(f"[TOOL] read_bom_excel called with {filepath}")
        # Llegir Excel saltant les primeres 6 files (metadata)
        # La fila 7 (√≠ndex 6) cont√© els headers
        df_bom = pd.read_excel(filepath, skiprows=7)
        
        # Filtrar nom√©s files amb MATERIAL (components reals)
        df_bom = df_bom[df_bom['MATERIAL'].notna()].copy()
        df_bom['MATERIAL'] = df_bom['MATERIAL'].astype(str).str.strip()
        df_bom['MATERIAL'] = df_bom['MATERIAL'].str.replace(r'\.0+$', '', regex=True)
        df_bom = df_bom[df_bom['MATERIAL'].str.match(r'^\d+$', na=False)].copy()
    
        if len(df_bom) == 0:
            return json.dumps([])
        
        # Netejar quantitats
        df_bom['Quantity'] = pd.to_numeric(df_bom['Quantity'], errors='coerce').fillna(1).astype(int)
        
        # Categoritzar components per tipus
        def get_category(desc):
            if pd.isna(desc):
                return 'Unknown'
            desc_lower = str(desc).lower()
            if 'diode' in desc_lower:
                if 'led' in desc_lower:
                    return 'LED'
                elif 'zener' in desc_lower:
                    return 'Diode-Zener'
                elif 'schottky' in desc_lower:
                    return 'Diode-Schottky'
                else:
                    return 'Diode'
            elif 'fuse' in desc_lower:
                return 'Fuse/Protection'
            elif 'cap' in desc_lower or 'capacitor' in desc_lower:
                if 'alum' in desc_lower or 'electrolytic' in desc_lower:
                    return 'Capacitor-Electrolytic'
                return 'Capacitor-Ceramic'
            elif 'res' in desc_lower or 'resistor' in desc_lower:
                return 'Resistor'
            elif 'ic ' in desc_lower or 'mcu' in desc_lower:
                return 'IC'
            elif 'relay' in desc_lower or 'rel√©' in desc_lower:
                return 'Relay'
            elif 'trans' in desc_lower:
                if 'former' in desc_lower:
                    return 'Transformer'
                return 'Transistor'
            elif 'switch' in desc_lower:
                return 'Switch'
            elif 'buzzer' in desc_lower:
                return 'Buzzer'
            else:
                return 'Other'
        
        df_bom['Category'] = df_bom['Description'].apply(get_category)
        
        # Crear llista de ComponentToQuote
        components = []
        for _, row in df_bom.iterrows():
            # Obtenir manufacturer si existeix, sin√≥ "Unknown"
            manufacturer = "Unknown"
            if 'Manufacturer' in df_bom.columns and pd.notna(row.get('Manufacturer')):
                manufacturer = str(row['Manufacturer'])
            elif 'Fabricant' in df_bom.columns and pd.notna(row.get('Fabricant')):
                manufacturer = str(row['Fabricant'])
            
            component = ComponentToQuote(
                codi_JCM=str(row['MATERIAL']),
                description=str(row['Description']) if pd.notna(row['Description']) else "No description",
                quantity=int(row['Quantity']),
                manufacturer=manufacturer,
                category=row['Category'],
                part_number=str(row['Part Number'])
            )
            components.append(component)
        
        # Retornar com a JSON string (serialitzant els objectes Pydantic)
        return json.dumps([comp.model_dump() for comp in components], indent=2, ensure_ascii=False)
        
    except FileNotFoundError:
        return json.dumps({"error": f"No s'ha trobat el fitxer {filepath}"})
    except KeyError as e:
        return json.dumps({"error": f"No s'ha trobat la columna esperada: {e}"})
    except Exception as e:
        return json.dumps({"error": f"ERROR llegint el fitxer: {str(e)}"})

print("‚úÖ Tool 'read_bom_excel' creada (retorna ComponentToQuote en format JSON)!")
````

### 6.1 Authenticate Gmail

Primer creem una funci√≥ per autenticar amb Gmail:

````python
import os
from google.auth.transport.requests import Request
from google.oauth2.credentials import Credentials
from google_auth_oauthlib.flow import InstalledAppFlow
from googleapiclient.discovery import build

SCOPES = [
    "https://www.googleapis.com/auth/gmail.send",
    "https://www.googleapis.com/auth/gmail.readonly",
    "https://www.googleapis.com/auth/gmail.modify",
]

def authenticate_gmail():
    """
    Autentica Gmail API. Si el token est√† revocat/expirat (invalid_grant),
    esborra token.json i for√ßa re-login.
    """
    creds = None
    token_path = "token.json"
    credentials_path = "credentials.json"

    # 1) Carrega token existent si hi √©s
    if os.path.exists(token_path):
        creds = Credentials.from_authorized_user_file(token_path, SCOPES)

    # 2) Si no hi ha creds o no s√≥n v√†lides ‚Üí refresca o re-login
    if not creds or not creds.valid:
        # Si hi ha refresh_token, intenta refresh
        if creds and creds.expired and creds.refresh_token:
            try:
                print("üîÑ Renovant token...")
                creds.refresh(Request())
            except Exception as e:
                # Cas t√≠pic: invalid_grant
                print(f"‚ö†Ô∏è Refresh fallat: {e}")
                print("üßπ Esborrant token.json i for√ßant re-autenticaci√≥...")
                if os.path.exists(token_path):
                    os.remove(token_path)
                creds = None

        # Si no s'ha pogut refrescar (o no hi havia refresh_token), re-login
        if not creds or not creds.valid:
            if not os.path.exists(credentials_path):
                raise FileNotFoundError(
                    "‚ùå No s'ha trobat credentials.json!\n"
                    "Has de descarregar-lo del Google Cloud Console i posar-lo a la carpeta del projecte."
                )

            print("üîê Primera autenticaci√≥ / Re-autenticaci√≥...")
            print("   S'obrir√† un navegador per autenticar-te.")
            flow = InstalledAppFlow.from_client_secrets_file(credentials_path, SCOPES)

            creds = flow.run_local_server(
                port=0,
                access_type="offline",
                prompt="consent"
            )

        # Guarda token nou
        with open(token_path, "w") as token_file:
            token_file.write(creds.to_json())
        print("‚úÖ Token guardat a token.json")

    # 3) Crea servei Gmail
    service = build("gmail", "v1", credentials=creds)

    # 4) Mostra usuari autenticat
    profile = service.users().getProfile(userId="me").execute()
    print(f"‚úÖ Autenticat com: {profile['emailAddress']}")

    return service
````

### 6.3 Funci√≥ Send Email

El fet d'enviar emails no fa falta que ho fem amb un agent, ho podem fer directament amb una funci√≥.

````python
import base64
from email.mime.text import MIMEText

def send_email_gmail(service, to_email: str, subject: str, body: str) -> str:
    message = MIMEText(body, _charset="utf-8")
    message["to"] = to_email
    message["subject"] = subject

    raw_message = base64.urlsafe_b64encode(message.as_bytes()).decode("utf-8")

    sent = service.users().messages().send(
        userId="me",
        body={"raw": raw_message}
    ).execute()

    return sent["id"]


print("‚úÖ Funci√≥ 'send_email_gmail' creada!")
````

### 6.4 Funci√≥ Fetch Emails

Funci√≥ per llegir emails rebuts:

````python
import re

def _decode_gmail_body_data(data: str) -> str:
    """Decodifica base64url -> text utf-8"""
    if not data:
        return ""
    return base64.urlsafe_b64decode(data.encode("utf-8")).decode("utf-8", errors="replace")


def _extract_body_from_payload(payload: dict) -> str:
    """
    Intenta extreure body real:
    1) text/plain
    2) text/html
    3) si no hi ha parts, mira body.data
    """
    if not payload:
        return ""

    # cas simple: body directe
    if payload.get("body", {}).get("data"):
        return _decode_gmail_body_data(payload["body"]["data"])

    parts = payload.get("parts", [])
    if not parts:
        return ""

    text_plain = None
    text_html = None

    def walk(parts_list):
        nonlocal text_plain, text_html
        for part in parts_list:
            mime = part.get("mimeType", "")
            body_data = part.get("body", {}).get("data")

            if mime == "text/plain" and body_data and not text_plain:
                text_plain = _decode_gmail_body_data(body_data)

            if mime == "text/html" and body_data and not text_html:
                text_html = _decode_gmail_body_data(body_data)

            # multipart dins multipart
            if "parts" in part:
                walk(part["parts"])

    walk(parts)

    return text_plain or text_html or ""


def fetch_emails_gmail(
    query: str = "newer_than:1d",
    max_results: int = 10,
    subject_contains: Optional[str] = None,
    from_contains: Optional[str] = None,
    only_replies: bool = False,
    limit: Optional[int] = None
) -> GmailInbox:
    """
    Llegeix emails i retorna GmailInbox (Pydantic) amb body real.
    Permet filtres extra en Python (m√©s fiables que nom√©s Gmail query).
    """
    global gmail_service

    results = gmail_service.users().messages().list(
        userId="me",
        q=query,
        maxResults=max_results
    ).execute()

    messages = results.get("messages", [])
    emails_out: List[GmailEmail] = []

    # Preparem regex (si toca)
    subject_re = re.compile(re.escape(subject_contains), re.IGNORECASE) if subject_contains else None
    from_re = re.compile(re.escape(from_contains), re.IGNORECASE) if from_contains else None

    for msg in messages:
        msg_id = msg["id"]

        msg_data = gmail_service.users().messages().get(
            userId="me",
            id=msg_id,
            format="full"
        ).execute()

        headers = msg_data.get("payload", {}).get("headers", [])
        subject = next((h["value"] for h in headers if h.get("name") == "Subject"), "") or ""
        from_email = next((h["value"] for h in headers if h.get("name") == "From"), "") or ""
        date = next((h["value"] for h in headers if h.get("name") == "Date"), "") or ""

        # ‚úÖ FILTRES HARD (Python)
        if subject_re and not subject_re.search(subject):
            continue
        if from_re and not from_re.search(from_email):
            continue
        if only_replies and not subject.lower().startswith("re:"):
            continue

        snippet = msg_data.get("snippet", "") or ""

        payload = msg_data.get("payload", {})
        body_real = _extract_body_from_payload(payload)

        # fallback si no hem trobat res
        if not body_real.strip():
            body_real = snippet

        emails_out.append(GmailEmail(
            message_id=msg_id,
            from_email=from_email,
            subject=subject,
            date=date,
            snippet=snippet,
            body=body_real
        ))

        # limit (per tallar aviat i anar m√©s r√†pid)
        if limit and len(emails_out) >= limit:
            break

    return GmailInbox(
        emails=emails_out,
        total_emails=len(emails_out)
    )

print("‚úÖ Funci√≥ 'fetch_emails_gmail' creada!")
````

## 9. Generaci√≥ de l'informe

Ara que tenim les dades extretes, analitzem la informaci√≥ rebuda i generem l'informe final.

### 9.1 An√†lisis d'ofertes

````python
def _build_quote_index(parsed: "ParsedQuotes") -> Dict[str, List["ParsedQuoteLine"]]:
    """
    Extreu totes les ofertes i les agrupa per codi.
    Retorna:
      quotes_by_codi[codi_JCM] = [ParsedQuoteLine, ParsedQuoteLine, ...]
    """
    quotes_by_codi: Dict[str, List["ParsedQuoteLine"]] = {}

    for resp in parsed.responses:
        for line in resp.lines:
            key = (line.codi_JCM or "").strip()
            if not key:
                continue
            quotes_by_codi.setdefault(key, []).append(line)

    return quotes_by_codi


def _select_best_offer(quotes: List["ParsedQuoteLine"]) -> Optional["ParsedQuoteLine"]:
    """
    Estrat√®gia simple: escollir el preu unitari m√©s baix (si existeix).
    Si unit_price √©s None ‚Üí no pot competir.
    """
    valid = [q for q in quotes if q.unit_price is not None]
    if not valid:
        return None
    return sorted(valid, key=lambda q: q.unit_price)[0]

def valuate_bom_best_offers(analysis: "BOMAnalysis", parsed: "ParsedQuotes") -> ValuatedBOM:
    """
    Extreu la millor oferta per cada component del BOM inicial

    """
    quotes_by_codi = _build_quote_index(parsed)

    lines_out: List[ValuatedLine] = []

    # per totals per prove√Ødor (nom√©s millor oferta)
    supplier_subtotals: Dict[str, float] = {}
    supplier_num_lines: Dict[str, int] = {}

    grand_total = 0.0
    total_missing = 0
    total_valuated = 0

    for comp in analysis.components_to_quote:
        codi = comp.codi_JCM.strip()
        quotes_for_this = quotes_by_codi.get(codi, [])

        quoted_by = sorted(list({q.supplier for q in quotes_for_this if q.supplier}))
        best = _select_best_offer(quotes_for_this)

        if not best:
            total_missing += 1
            lines_out.append(ValuatedLine(
                codi_JCM=codi,
                description=comp.description,
                manufacturer=comp.manufacturer,
                part_number=comp.part_number,
                quantity=comp.quantity,
                best_offer=None,
                line_total_eur=None,
                quoted_by=quoted_by,
                missing_quote=True
            ))
            continue

        line_total = float(best.unit_price) * int(comp.quantity)

        best_offer = BestOffer(
            supplier=best.supplier,
            unit_price=float(best.unit_price),
            currency=best.currency or "EUR",
            stock=getattr(best, "stock", None),
            stock_text=getattr(best, "stock_text", None),
            lead_time_days=getattr(best, "lead_time_days", None),
            lead_time_text=getattr(best, "lead_time_text", None),
            moq=getattr(best, "moq", None),
            notes=getattr(best, "notes", None),
        )

        lines_out.append(ValuatedLine(
            codi_JCM=codi,
            description=comp.description,
            manufacturer=comp.manufacturer,
            part_number=comp.part_number,
            quantity=comp.quantity,
            best_offer=best_offer,
            line_total_eur=line_total,
            quoted_by=quoted_by,
            missing_quote=False
        ))

        grand_total += line_total
        total_valuated += 1

        supplier_subtotals[best.supplier] = supplier_subtotals.get(best.supplier, 0.0) + line_total
        supplier_num_lines[best.supplier] = supplier_num_lines.get(best.supplier, 0) + 1

    totals_by_supplier = [
        SupplierTotal(supplier=s, subtotal_eur=round(sub, 4), num_lines=supplier_num_lines.get(s, 0))
        for s, sub in sorted(supplier_subtotals.items(), key=lambda x: x[0])
    ]

    return ValuatedBOM(
        lines=lines_out,
        total_lines=len(lines_out),
        total_valuated_lines=total_valuated,
        total_missing_lines=total_missing,
        grand_total_eur=round(grand_total, 4),
        totals_by_supplier=totals_by_supplier
    )

print("Rutines creades!!")
````
