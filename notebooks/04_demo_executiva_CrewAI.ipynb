## ğŸ” AutenticaciÃ³ Gmail


````python
import os
from google.auth.transport.requests import Request
from google.oauth2.credentials import Credentials
from google_auth_oauthlib.flow import InstalledAppFlow
from googleapiclient.discovery import build

SCOPES = [
    "https://www.googleapis.com/auth/gmail.send",
    "https://www.googleapis.com/auth/gmail.readonly",
    "https://www.googleapis.com/auth/gmail.modify",
]

def authenticate_gmail():
    """
    Autentica Gmail API. Si el token estÃ  revocat/expirat (invalid_grant),
    esborra token.json ià¦—à¦—a re-login.
    """
    creds = None
    token_path = "token.json"
    credentials_path = "credentials.json"

    if os.path.exists(token_path):
        creds = Credentials.from_authorized_user_file(token_path, SCOPES)

    if not creds or not creds.valid:
        if creds and creds.expired and creds.refresh_token:
            try:
                print("ğŸ”„ Renovant token...")
                creds.refresh(Request())
            except Exception as e:
                print(f"âš ï¸ Refresh fallat: {e}")
                print("ğŸ§¹ Esborrant token.json i forÃ§ant re-autenticaciÃ³...")
                if os.path.exists(token_path):
                    os.remove(token_path)
                creds = None

        if not creds or not creds.valid:
            if not os.path.exists(credentials_path):
                raise FileNotFoundError(
                    "âŒ No s'ha trobat credentials.json!\n"
                    "Has de descarregar-lo del Google Cloud Console i posar-lo a la carpeta del projecte."
                )

            print("ğŸ” Primera autenticaciÃ³ / Re-autenticaciÃ³...")
            print("   S'obrirÃ  un navegador per autenticar-te.")
            flow = InstalledAppFlow.from_client_secrets_file(credentials_path, SCOPES)

            creds = flow.run_local_server(
                port=0,
                access_type="offline",
                prompt="consent"
            )

        with open(token_path, "w") as token_file:
            token_file.write(creds.to_json())
        print("âœ… Token guardat a token.json")

    service = build("gmail", "v1", credentials=creds)

    profile = service.users().getProfile(userId="me").execute()
    print(f"âœ… Autenticat com: {profile['emailAddress']}")

    return service

# Autenticar
try:
    gmail_service = authenticate_gmail()
    print("\nâœ… Gmail API configurada correctament!")
except FileNotFoundError as e:
    print(f"\nâŒ Error: {e}")
except Exception as e:
    print(f"\nâŒ Error inesperat: {e}")
````

````python
# ============================================================================
# TOOLS (ConfiguraciÃ³ Part 3)
# ============================================================================

from crewai.tools import tool
import pandas as pd
import json
import re
import base64
import time
from datetime import datetime, timedelta
from typing import List, Optional, Dict
from email.mime.text import MIMEText

# â”€â”€â”€ TOOL 1: Read BOM Excel â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

@tool("Read BOM Excel")
def read_bom_excel(filepath: str) -> str:
    """
    Llegeix un fitxer Excel que contÃ© un Bill of Materials (BOM) de JCM Technologies.
    Processa el format especÃ­fic de JCM amb metadata inicial i headers a la fila 7.
    Retorna una llista de ComponentToQuote en format JSON.

    Args:
        filepath: Ruta al fitxer Excel (ex: 'data/Bill_of_Materials-TESTBOM.xlsx')

    Returns:
        JSON string amb llista de ComponentToQuote objects
    """
    try:
        print(f"[TOOL] read_bom_excel called with {filepath}")
        df_bom = pd.read_excel(filepath, skiprows=7)

        df_bom = df_bom[df_bom['MATERIAL'].notna()].copy()
        df_bom['MATERIAL'] = df_bom['MATERIAL'].astype(str).str.strip()
        df_bom['MATERIAL'] = df_bom['MATERIAL'].str.replace(r'\.0+$', '', regex=True)
        df_bom = df_bom[df_bom['MATERIAL'].str.match(r'^\d+$', na=False)].copy()

        if len(df_bom) == 0:
            return json.dumps([])

        df_bom['Quantity'] = pd.to_numeric(df_bom['Quantity'], errors='coerce').fillna(1).astype(int)

        def get_category(desc):
            if pd.isna(desc):
                return 'Unknown'
            desc_lower = str(desc).lower()
            if 'diode' in desc_lower:
                if 'led' in desc_lower:
                    return 'LED'
                elif 'zener' in desc_lower:
                    return 'Diode-Zener'
                elif 'schottky' in desc_lower:
                    return 'Diode-Schottky'
                else:
                    return 'Diode'
            elif 'fuse' in desc_lower:
                return 'Fuse/Protection'
            elif 'cap' in desc_lower or 'capacitor' in desc_lower:
                if 'alum' in desc_lower or 'electrolytic' in desc_lower:
                    return 'Capacitor-Electrolytic'
                return 'Capacitor-Ceramic'
            elif 'res' in desc_lower or 'resistor' in desc_lower:
                return 'Resistor'
            elif 'ic ' in desc_lower or 'mcu' in desc_lower:
                return 'IC'
            elif 'relay' in desc_lower or 'relÃ©' in desc_lower:
                return 'Relay'
            elif 'trans' in desc_lower:
                if 'former' in desc_lower:
                    return 'Transformer'
                return 'Transistor'
            elif 'switch' in desc_lower:
                return 'Switch'
            elif 'buzzer' in desc_lower:
                return 'Buzzer'
            else:
                return 'Other'

        df_bom['Category'] = df_bom['Description'].apply(get_category)

        components = []
        for _, row in df_bom.iterrows():
            manufacturer = "Unknown"
            if 'Manufacturer' in df_bom.columns and pd.notna(row.get('Manufacturer')):
                manufacturer = str(row['Manufacturer'])
            elif 'Fabricant' in df_bom.columns and pd.notna(row.get('Fabricant')):
                manufacturer = str(row['Fabricant'])

            component = ComponentToQuote(
                codi_JCM=str(row['MATERIAL']),
                description=str(row['Description']) if pd.notna(row['Description']) else "No description",
                quantity=int(row['Quantity']),
                manufacturer=manufacturer,
                category=row['Category'],
                part_number=str(row['Part Number'])
            )
            components.append(component)

        return json.dumps([comp.model_dump() for comp in components], indent=2, ensure_ascii=False)

    except FileNotFoundError:
        return json.dumps({"error": f"No s'ha trobat el fitxer {filepath}"})
    except KeyError as e:
        return json.dumps({"error": f"No s'ha trobat la columna esperada: {e}"})
    except Exception as e:
        return json.dumps({"error": f"ERROR llegint el fitxer: {str(e)}"})

print("âœ… Tool 'read_bom_excel' creada!")

# â”€â”€â”€ TOOL 2: Format RFQ Email â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

@tool("Format RFQ Email")
def format_rfq_email(
    supplier_name: str,
    to_email: str,
    components_list: str,
    rfq_number: str,
    project_name: Optional[str] = None,
    from_email: str = "compres@jcm-tech.com"
) -> Dict[str, str]:
    """
    Retorna un email RFQ com un objecte estructurat:
    { "subject": "...", "body": "..." }
    """
    supplier_name = (supplier_name or "").strip()
    to_email = (to_email or "").strip()
    rfq_number = (rfq_number or "").strip()
    components_list = (components_list or "").strip()

    if not supplier_name or not rfq_number:
        return {"error": "Falta supplier_name o rfq_number."}
    if not components_list:
        return {"error": "components_list buit. No es pot generar RFQ."}

    today = datetime.now().strftime("%d/%m/%Y")
    deadline = (datetime.now() + timedelta(days=7)).strftime("%d/%m/%Y")

    subject = f"SolÂ·licitud de cotitza #{rfq_number} - Components electrÃ²nics"
    if project_name:
        subject = f"SolÂ·licitud de cotitza #{rfq_number} - {project_name} - Components electrÃ²nics"

    body = f"""Benvolgut/da equip de {supplier_name},

Des de JCM Technologies, SAU, volem solÂ·licitar-vos una cotitza per als
segons components electrÃ²nics.

COMPONENTS SOLÂ·LICITATS:
â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
{components_list}
â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

INFORMACIÃ“ ADDICIONAL:
- ReferÃ¨ncia RFQ: {rfq_number}
- Data lÃ­mit resposta: {deadline}
- Lliurament preferit: Vic, Barcelona, Espanya
- Condicions de pagament: 30 dies data factura

Us agrairÃ­em que ens proporcionÃ©ssiu:
âœ“ Preu unitari per component
âœ“ Disponibilitat de stock actual
âœ“ Temps de lliurament estimat
âœ“ MOQ (Minimum Order Quantity) si aplica
âœ“ Validesa de la cotitza

Qualsevol dubte sobre especificacions tÃ¨cniques, no dubteu en contactar-nos.

Atentament,

Departament de Compres
JCM Technologies, SAU
Email: {from_email}"""

    return {
        "from_email": from_email,
        "to_email": to_email,
        "subject": subject,
        "body": body
    }

print("âœ… Tool 'format_rfq_email' creada!")

# â”€â”€â”€ FUNCIÃ“: Send Email (directa, no tool) â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

def send_email_gmail(service, to_email: str, subject: str, body: str) -> str:
    message = MIMEText(body, _charset="utf-8")
    message["to"] = to_email
    message["subject"] = subject

    raw_message = base64.urlsafe_b64encode(message.as_bytes()).decode("utf-8")

    sent = service.users().messages().send(
        userId="me",
        body={"raw": raw_message}
    ).execute()

    return sent["id"]

print("âœ… FunciÃ³ 'send_email_gmail' creada!")

# â”€â”€â”€ FUNCIÃ“: Fetch Emails (directa, retorna GmailInbox Pydantic) â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

def _decode_gmail_body_data(data: str) -> str:
    """Decodifica base64url -> text utf-8"""
    if not data:
        return ""
    return base64.urlsafe_b64decode(data.encode("utf-8")).decode("utf-8", errors="replace")


def _extract_body_from_payload(payload: dict) -> str:
    if not payload:
        return ""

    if payload.get("body", {}).get("data"):
        return _decode_gmail_body_data(payload["body"]["data"])

    parts = payload.get("parts", [])
    if not parts:
        return ""

    text_plain = None
    text_html = None

    def walk(parts_list):
        nonlocal text_plain, text_html
        for part in parts_list:
            mime = part.get("mimeType", "")
            body_data = part.get("body", {}).get("data")

            if mime == "text/plain" and body_data and not text_plain:
                text_plain = _decode_gmail_body_data(body_data)

            if mime == "text/html" and body_data and not text_html:
                text_html = _decode_gmail_body_data(body_data)

            if "parts" in part:
                walk(part["parts"])

    walk(parts)

    return text_plain or text_html or ""


def fetch_emails_gmail(
    query: str = "newer_than:1d",
    max_results: int = 10,
    subject_contains: Optional[str] = None,
    from_contains: Optional[str] = None,
    only_replies: bool = False,
    limit: Optional[int] = None
) -> GmailInbox:
    """
    Llegeix emails i retorna GmailInbox (Pydantic) amb body real.
    """
    global gmail_service

    results = gmail_service.users().messages().list(
        userId="me",
        q=query,
        maxResults=max_results
    ).execute()

    messages = results.get("messages", [])
    emails_out: List[GmailEmail] = []

    subject_re = re.compile(re.escape(subject_contains), re.IGNORECASE) if subject_contains else None
    from_re = re.compile(re.escape(from_contains), re.IGNORECASE) if from_contains else None

    for msg in messages:
        msg_id = msg["id"]

        msg_data = gmail_service.users().messages().get(
            userId="me",
            id=msg_id,
            format="full"
        ).execute()

        headers = msg_data.get("payload", {}).get("headers", [])
        subject = next((h["value"] for h in headers if h.get("name") == "Subject"), "") or ""
        from_email = next((h["value"] for h in headers if h.get("name") == "From"), "") or ""
        date = next((h["value"] for h in headers if h.get("name") == "Date"), "") or ""

        if subject_re and not subject_re.search(subject):
            continue
        if from_re and not from_re.search(from_email):
            continue
        if only_replies and not subject.lower().startswith("re:"):
            continue

        snippet = msg_data.get("snippet", "") or ""

        payload = msg_data.get("payload", {})
        body_real = _extract_body_from_payload(payload)

        if not body_real.strip():
            body_real = snippet

        emails_out.append(GmailEmail(
            message_id=msg_id,
            from_email=from_email,
            subject=subject,
            date=date,
            snippet=snippet,
            body=body_real
        ))

        if limit and len(emails_out) >= limit:
            break

    return GmailInbox(
        emails=emails_out,
        total_emails=len(emails_out)
    )

print("âœ… FunciÃ³ 'fetch_emails_gmail' creada!")
````

````python
# ============================================================================
# FASE 2: LLEGIR RESPOSTES â†’ PARSEJAR â†’ VALORAR â†’ GENERAR INFORME
# ============================================================================

from crewai import Crew, Process
from datetime import datetime
from typing import Dict

print("ğŸ” PROCESSANT RESPOSTES")
print("=" * 80)

# â”€â”€ Pas 1: Llegir emails de resposta â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

print("\nğŸ“¥ Llegint respostes dels voluntaris...\n")

inbox = fetch_emails_gmail(
    query="newer_than:1d",
    max_results=30,
    subject_contains="RFQ-DEMO-001",
    only_replies=True,
    limit=3
)

print(f"ğŸ“¥ GmailInbox: {inbox.total_emails} correus")
for i, e in enumerate(inbox.emails, start=1):
    print(f"   [{i}] {e.subject} | {e.from_email} | {e.date}")

# â”€â”€ Pas 2: Email Parser â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

print("\nâ³ Executant Agent 4: Email Parser...\n")
print("=" * 80)

task_parse_emails.description += f"""
    EMAILS (INPUT): {inbox.model_dump_json(indent=2)}"""

crew_parse = Crew(
    agents=[agent_email_parser],
    tasks=[task_parse_emails],
    process=Process.sequential,
    verbose=False,
    max_iter=5,
    max_rpm=60
)

print(f"Hora inici: {datetime.now().strftime('%H:%M:%S')}")

result_parse = crew_parse.kickoff()

parsed_data: ParsedQuotes = result_parse.pydantic

print(f"\nâœ… Parsing completat!")
print(f"Hora fi: {datetime.now().strftime('%H:%M:%S')}")
print(f"   Total respostes parsejades: {parsed_data.total_responses}")

for i, r in enumerate(parsed_data.responses, start=1):
    print(f"\n   [{i}] {r.supplier} â€” {len(r.lines)} lÃ­nies")
    for line in r.lines:
        print(f"       {line.codi_JCM} | {line.unit_price} {line.currency} | "
              f"stock={line.stock_text} | lead={line.lead_time_text}")

# â”€â”€ Pas 3: ValoraciÃ³ (best offers) â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

print("\nâ³ Calculant millors ofertes...\n")

def _build_quote_index(parsed: ParsedQuotes) -> Dict[str, list]:
    quotes_by_codi: Dict[str, list] = {}
    for resp in parsed.responses:
        for line in resp.lines:
            key = (line.codi_JCM or "").strip()
            if not key:
                continue
            quotes_by_codi.setdefault(key, []).append(line)
    return quotes_by_codi

def _select_best_offer(quotes):
    valid = [q for q in quotes if q.unit_price is not None]
    if not valid:
        return None
    return sorted(valid, key=lambda q: q.unit_price)[0]

def valuate_bom_best_offers(analysis: BOMAnalysis, parsed: ParsedQuotes) -> ValuatedBOM:
    quotes_by_codi = _build_quote_index(parsed)

    lines_out = []
    supplier_subtotals: Dict[str, float] = {}
    supplier_num_lines: Dict[str, int] = {}
    grand_total = 0.0
    total_missing = 0
    total_valuated = 0

    for comp in analysis.components_to_quote:
        codi = comp.codi_JCM.strip()
        quotes_for_this = quotes_by_codi.get(codi, [])
        quoted_by = sorted(list({q.supplier for q in quotes_for_this if q.supplier}))
        best = _select_best_offer(quotes_for_this)

        if not best:
            total_missing += 1
            lines_out.append(ValuatedLine(
                codi_JCM=codi,
                description=comp.description,
                manufacturer=comp.manufacturer,
                part_number=comp.part_number,
                quantity=comp.quantity,
                best_offer=None,
                line_total_eur=None,
                quoted_by=quoted_by,
                missing_quote=True
            ))
            continue

        line_total = float(best.unit_price) * int(comp.quantity)

        best_offer = BestOffer(
            supplier=best.supplier,
            unit_price=float(best.unit_price),
            currency=best.currency or "EUR",
            stock=getattr(best, "stock", None),
            stock_text=getattr(best, "stock_text", None),
            lead_time_days=getattr(best, "lead_time_days", None),
            lead_time_text=getattr(best, "lead_time_text", None),
            moq=getattr(best, "moq", None),
            notes=getattr(best, "notes", None),
        )

        lines_out.append(ValuatedLine(
            codi_JCM=codi,
            description=comp.description,
            manufacturer=comp.manufacturer,
            part_number=comp.part_number,
            quantity=comp.quantity,
            best_offer=best_offer,
            line_total_eur=line_total,
            quoted_by=quoted_by,
            missing_quote=False
        ))

        grand_total += line_total
        total_valuated += 1
        supplier_subtotals[best.supplier] = supplier_subtotals.get(best.supplier, 0.0) + line_total
        supplier_num_lines[best.supplier] = supplier_num_lines.get(best.supplier, 0) + 1

    totals_by_supplier = [
        SupplierTotal(supplier=s, subtotal_eur=round(sub, 4), num_lines=supplier_num_lines.get(s, 0))
        for s, sub in sorted(supplier_subtotals.items(), key=lambda x: x[0])
    ]

    return ValuatedBOM(
        lines=lines_out,
        total_lines=len(lines_out),
        total_valuated_lines=total_valuated,
        total_missing_lines=total_missing,
        grand_total_eur=round(grand_total, 4),
        totals_by_supplier=totals_by_supplier
    )

valuated = valuate_bom_best_offers(analysis, parsed_data)

print(f"ğŸ“¦ ESCANDALL VALORAT:")
print(f"   Total lÃ­nies BOM:        {valuated.total_lines}")
print(f"   LÃ­nies amb oferta:       {valuated.total_valuated_lines}")
print(f"   LÃ­nies sense cotitzar:   {valuated.total_missing_lines}")
print(f"   TOTAL GLOBAL (â‚¬):        {valuated.grand_total_eur:.4f}")
print(f"\nğŸ“Œ Totals per proveÃ¯dor:")
for t in valuated.totals_by_supplier:
    print(f"   - {t.supplier:10s} -> {t.subtotal_eur:10.4f} â‚¬  ({t.num_lines} lÃ­nies)")

# â”€â”€ Pas 4: Report Generator â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

print("\nâ³ Executant Agent 5: Report Generator...\n")
print("=" * 80)

task_generate_report.description += f"""

DADES DE COTITZA (ValuatedBOM):
{valuated.model_dump_json(indent=2)}

DADES PARSEADES COMPLETES (ParsedQuotes):
{parsed_data.model_dump_json(indent=2)}
"""

crew_report = Crew(
    agents=[agent_report_generator],
    tasks=[task_generate_report],
    process=Process.sequential,
    verbose=False,
    max_iter=3,
    max_rpm=60
)

print(f"Hora inici: {datetime.now().strftime('%H:%M:%S')}")

result_report = crew_report.kickoff()

print(f"\nâœ… INFORME GENERAT!")
print(f"Hora fi: {datetime.now().strftime('%H:%M:%S')}")
print("=" * 80)
````
